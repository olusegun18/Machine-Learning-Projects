```python
from sklearn.ensemble import GradientBoostingRegressor
from catboost import CatBoostRegressor,Pool
from sklearn.model_selection import KFold, cross_val_score

params = {'n_estimators':100, 'max_depth':4, 'min_samples_split':2, 'learning_rate': 0.05}
cat = CatBoostRegressor(silent=True, iterations=100, learning_rate=0.1, depth =2, loss_function = 'RMSE')
gbr_model = GradientBoostingRegressor(**params)
gbr_model.fit(X_train,y_train)
cat_model = cat.fit(X_train,y_train)
```

```python
from xgboost import XGBRegressor

xgb = XGBRegressor(n_estimators=10000, learning_rate=0.01)
xgb_model = xgb.fit(X_train, y_train, 
             early_stopping_rounds=5, 
             eval_set=[(X_train, y_train)],
             verbose=False)
xgb.score(X_train,y_train)
```

```python
xgb_pred =xgb_model.predict(X_test)
score = np.sqrt(mean_squared_error(xgb_pred,y_test))
print("cat: {:.4f}".format(score))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

my_pipeline = Pipeline(steps=[('preprocessor', SimpleImputer()),
                              ('model', RandomForestRegressor(n_estimators=50,
                                                              random_state=0))
                             ])
```

```python
from sklearn.model_selection import cross_val_score

# Multiply by -1 since sklearn calculates *negative* MAE
scores = -1 * cross_val_score(my_pipeline, X, y,
                              cv=5,
                              scoring='neg_mean_squared_error')

print("MSE scores:\n", scores)
print("Average MSE score (across experiments): {:.5f}".format(scores.mean()))
```
