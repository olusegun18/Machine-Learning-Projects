```python
X = data.drop(['Item_Store_Returns'], axis = 1)
y = data['Item_Store_Returns']
```

```python
# from sklearn.preprocessing import MinMaxScaler, StandardScaler
# scaler= MinMaxScaler()
# X = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=0)
```

```python
from catboost import CatBoostRegressor,Pool
from sklearn.metrics import mean_squared_error
```

```python
# cat_index = [0,1,4,5,6]
cat_index = [1,2,5,6,7]

# cat = CatBoostRegressor(silent=True, iterations=100, learning_rate=0.05)
cat = CatBoostRegressor(silent=True, iterations=100, learning_rate=0.05, depth =10, loss_function = 'RMSE')
cat_model = cat.fit(Pool(X_train,y_train,cat_features = cat_index))
# cat_model.score(X_train,y_train)
c_pred = cat_model.predict(X_test)
cat_score = np.sqrt(mean_squared_error(c_pred,y_test))
print("cat: {:.4f}".format(cat_score))
```
