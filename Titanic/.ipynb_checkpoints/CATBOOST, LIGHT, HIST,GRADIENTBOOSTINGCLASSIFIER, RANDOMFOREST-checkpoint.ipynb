```python
## RandomForestBagging

bags=10
seed = 1
r_bagged_predictions = np.zeros(test.shape[0])

for n in range(0,bags):
    r_class.set_params(random_state = seed + n)
    r_class.fit(X,y)
    r_preds = r_class.predict_proba(test)[:,-1]
    r_bagged_predictions +=r_preds
r_bagged_predictions/=bags
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_auc_score


features = train.select_dtypes(exclude='object').columns

def metric(y,pred):
    return roc_auc_score(y,pred, labels = [0,1])

n_skf = 5
kf = StratifiedKFold(n_skf)
seed = 2020

params = {'n_estimators':4000,
          'criterion':'gini',
          'max_depth':10
         }

          
```

```python
score_list = []
score = 0
test_oofs = []
          
          
          
for i, (tr_idx, vr_idx) in enumerate(kf.split(X,y)):
          xtrain, ytrain = X.loc[tr_idx, features], y.loc[tr_idx]
          xval,yval = X.loc[vr_idx, features], y.loc[vr_idx]
          r_model = RandomForestClassifier(**params)
          r_model.fit(xtrain, ytrain)
          
          p = r_model.predict_proba(xval)[:,1]
          sc = metric(yval,p)
          score_list.append(sc)
          score += sc/n_skf
          
          pred  = r_model.predict_proba(test)[:,1]
          test_oofs.append(pred)
          
          
          print('Fold {} : {}'.format(i,sc))
          
          
print('Avg log: ', score)
```
